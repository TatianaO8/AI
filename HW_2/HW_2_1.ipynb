# HW 2 (Problem 1)

Description: 

Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that:

p1 is a minimum of f1,
p2 is a maximum of f2,
p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

##Minimum

```python
def f1(x, y):
    return x ** 2 + y ** 2 
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary', alpha=0.5)
ax.scatter(0,0,0, c='green')
ax.text(2,-2,0, "p1", color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('f1')
```

```python
plt.contour(X, Y, Z, 10, colors='black')
plt.scatter(0,0, c='green')
plt.show()
```

```python
plt.contourf(X, Y, Z, 50, cmap='RdGy')
plt.colorbar();
```

##Maximum

```python
def f2(x, y):
    return -(x ** 2 + y ** 2)
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary', alpha=0.5)
ax.scatter(0,0,0, c='blue')
ax.text(.5,1,5, "p2", color='blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.set_title('f2')
```

```python
plt.contour(X, Y, Z, colors='black')
plt.scatter(0,0, c='blue')
plt.show()
```

```python
plt.contourf(X, Y, Z, 50, cmap='RdGy')
plt.colorbar();
```

##Neither a min or max

```python
def f3(x, y):
    return x ** 3 + y ** 2
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary', alpha=0.5)
ax.scatter(0,0,0, c='red')
ax.text(0.5, 0.5, 0, "p3", c="red")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.set_title('f3')
```

```python
plt.contour(X, Y, Z, 50, colors='black')
plt.scatter(0,0, c='red')
plt.show()
```

```python
plt.contourf(X, Y, Z, 50, cmap='RdGy')
plt.colorbar();
```
